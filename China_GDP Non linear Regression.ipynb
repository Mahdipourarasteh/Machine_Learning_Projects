# Non Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Some commnn funktion's

## Linear funktion

```python
x= np.arange(-5,5,0.1)

y= 2 * x + 3
y_noise = 2 * np.random.normal(size=x.size)   #in normal methode the most of random numbers are between -1 and 1
ydata = y + y_noise

plt.plot(x, ydata, 'bo')
plt.plot(x,y, 'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

## cubic funktion

```python
x= np.arange(-5,5,0.1)

y= 1*(x**3) + 1*(x**2) + 1*x + 3
y_noise = 20 * np.random.normal(size=x.size)
ydata = y + y_noise
plt.plot(x, ydata, 'bo')
plt.plot(x,y,'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

## Quadratic funktion

```python
x= np.arange(-5,5,0.1)

y= np.power(x,2)
y_noise = 2 * np.random.normal(size=x.size)
ydata = y + y_noise
plt.plot(x, ydata, 'bo')
plt.plot(x,y,'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

## Exponential funktion

```python
x= np.arange(-5,5,0.1)

y= np.exp(x)

plt.plot(x,y)
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

## Logarithmic funktion

```python
x= np.arange(-5,5,0.1)

y= np.log(x)

plt.plot(x,y)
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

## Sigmoidal / logistic funktion

```python
x= np.arange(-5,5,0.1)

y= 1-4/(1+np.power(3, x-2))

plt.plot(x,y)
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

# Non Linear Regression

```python
import numpy as np
import pandas as pd

df= pd.read_csv('china_gdp.csv')
df.head()
```

## plotting the Dataset

```python
plt.figure(figsize=(8,5))
x_data, y_data = (df["Year"].values, df["Value"].values)
plt.plot(x_data, y_data, 'ro')
plt.xlabel('GDP')
plt.ylabel('Year')
plt.show()
```

## Choosing a model

```python
x= np.arange(-5,5,0.1)
y= 1/(1+np.exp(-x))  #sigmoidal or logistic funktion would be good for this economic data

plt.plot(x,y)
plt.ylabel('Dependend Variable')
plt.xlabel('Independend Variable')
plt.show()
```

## Building the model

```python
def sigmoid(x, Beta_1, Beta_2):
    y = 1 / (1 + np.exp(-Beta_1*(x-Beta_2)))
    return y
```

```python
beta_1 = 0.1
beta_2 = 1990

Y_pred = sigmoid(x_data, beta_1, beta_2)

plt.plot(x_data, Y_pred*15000000000000.)
plt.plot(x_data, y_data, 'ro')
```

## Normalize the input data

```python
xdata = x_data/ max(x_data)
ydata = y_data/ max(y_data)
```

## Optimize the model with curve_fit of scipy.otimize library

```python
from scipy.optimize import curve_fit    ## more details in documentation ot scipy.optimize curve_fit
popt, pcov = curve_fit(sigmoid, xdata, ydata)
print(" beta_1 = %f, beta_2 = %f" % (popt[0], popt[1]))
```

```python
x = np.linspace(1960, 2015, 55)   #55 years
x= x/max(x)
plt.figure(figsize=(8,5))
y= sigmoid(x, *popt)     # * send both of popt[0] and popt[1]
plt.plot(xdata, ydata, 'ro', label= 'data')
plt.plot(x,y, linewidth= 3, label= 'fit')
plt.legend(loc='best')
plt.xlabel('GDP')
plt.ylabel('Year')
plt.show()
```
