```python
from twelvedata import TDClient

def get_price(Symbol, Interval, n_interval):
    # Initialize client - apikey parameter is requiered
    td = TDClient(apikey="88d88fb76bc6449082320db208700469")

    # Construct the necessary time series
    ts = td.time_series(
        symbol= Symbol,
        interval= Interval,
        outputsize= n_interval,
        #timezone="America/New_York",
    )

    # Returns pandas.DataFrame
    df= ts.as_pandas()
    return df

dataset = get_price('GSPC', '1day', 5000)  #BTC/USD
dataset
```

## preparing the data for the prophet model

```python
dataset['date'] = dataset.index
df = dataset[['date','close']]
df
```

```python
df.dtypes
```

```python
df = df.sort_values(by = 'date')
df
```

```python
df.reset_index(inplace=True, drop=True)
df
```

```python
df.columns = ['ds', 'y']
df
```

```python
# Split the data into testing and training datasets
train , test = df[:-29], df[-29:]
```

```python
train
```

```python
test.head(10)
```

```python
m = Prophet()
m.fit(train)
```

```python
future = m.make_future_dataframe(periods=42)    
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
from prophet.plot import plot_plotly, plot_components_plotly

# Ploting the forecasted data

plot_plotly(m, forecast)
```

```python
from prophet.plot import add_changepoints_to_plot

# This particular feature helps us identify trend changes that are infered by the model

fig = m.plot(forecast)
a = add_changepoints_to_plot(fig.gca(), m, forecast)
```

## add missing dates to test set and evaluating the model

```python
#print(str(test.iloc[-1]['ds'].date()))
test_ = test.set_index('ds')
###print(test_)
test_ = test_.reindex(pd.date_range(str(test.iloc[0]['ds'].date()), str(test.iloc[-1]['ds'].date())))
test_ = test_.interpolate()  #fill the missing valius with linear interpolation methode
test_.head(10)
```

```python
first= int(forecast[forecast['ds'] == str(test.iloc[0]['ds'].date())].index.values)
last= int(forecast[forecast['ds'] == str(test.iloc[-1]['ds'].date())].index.values)
#print(forecast[first :last+1]['ds'])
score = r2_score(test_['y'], forecast[first :last+1]['yhat'])

print('R-Sqaure score is {}'.format(score))
```
