# Multiple Regression

## Import packages

```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
%matplotlib inline
```

## read the data

```python
df= pd.read_csv("FuelConsumption.csv")
df.head()
```

```python
cdf= df[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_CITY','FUELCONSUMPTION_HWY','FUELCONSUMPTION_COMB','CO2EMISSIONS']]
cdf.head()
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, constrained_layout= True, figsize=(9,3))

ax1.scatter(cdf.ENGINESIZE, cdf.CO2EMISSIONS, color='blue')
ax1.set_xlabel("ENGINESIZE")
ax1.set_ylabel("CO2EMISSIONS")

ax2.scatter(cdf.CYLINDERS, cdf.CO2EMISSIONS, color='blue')
ax2.set_xlabel("CYLINDERS")
ax2.set_ylabel("CO2EMISSIONS")

fig, (ax3, ax4, ax5) = plt.subplots(1,3, constrained_layout= True, figsize=(16,4))

ax3.scatter(cdf.FUELCONSUMPTION_CITY, cdf.CO2EMISSIONS, color='blue')
ax3.set_xlabel("FUELCONSUMPTION_CITY")
ax3.set_ylabel("CO2EMISSIONS")

ax4.scatter(cdf.FUELCONSUMPTION_HWY, cdf.CO2EMISSIONS, color='blue')
ax4.set_xlabel("FUELCONSUMPTION_HWY")
ax4.set_ylabel("CO2EMISSIONS")

ax5.scatter(cdf.FUELCONSUMPTION_COMB, cdf.CO2EMISSIONS, color='blue')
ax5.set_xlabel("FUELCONSUMPTION_COMB")
ax5.set_ylabel("CO2EMISSIONS")

plt.show()
```

## creating the train and test dataset

```python
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]
print(len(train))
print(len(test))
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, constrained_layout= True, figsize=(16,4))

ax1.scatter(train.ENGINESIZE, train.CO2EMISSIONS, color='blue')
ax1.scatter(test.ENGINESIZE, test.CO2EMISSIONS, color='red')
ax1.set_xlabel("ENGINESIZE")
ax1.set_ylabel("CO2EMISSIONS")

ax2.scatter(train.CYLINDERS, train.CO2EMISSIONS, color='blue')
ax2.scatter(test.CYLINDERS, test.CO2EMISSIONS, color='red')
ax2.set_xlabel("CYLINDERS")
ax2.set_ylabel("CO2EMISSIONS")

ax3.scatter(train.FUELCONSUMPTION_COMB, train.CO2EMISSIONS, color='blue')
ax3.scatter(test.FUELCONSUMPTION_COMB, test.CO2EMISSIONS, color='red')
ax3.set_xlabel("FUELCONSUMPTION_COMB")
ax3.set_ylabel("CO2EMISSIONS")

plt.show()
```

## multiple regression modeling

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
x= np.asanyarray(train[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']])
y= np.asanyarray(train[['CO2EMISSIONS']])
regr.fit(x,y)
print(regr)
print('Coefficients: ', regr.coef_)
print('Intercept_: ', regr.intercept_)
```

## Prediction

```python
y_hat= regr.predict(test[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']])
x= np.asanyarray(test[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']])
y= np.asanyarray(test[['CO2EMISSIONS']])
print("Residual sum of squares: %.2f" % np.mean((y_hat-y)**2))

print('Variance score: %.2f' % regr.score(x,y))
```
